```python
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Revision of FIX NYC-PROPERTY-SALES.xlsx")


def get_sale_price_by_borough(borough):
    """
    Hitung jumlah penjualan dan revenue per borough.

    Args:
        borough: Kode borough.

    Returns:
        tuple: Tuple berisi jumlah penjualan dan revenue per borough.
    """
    boroughes = df.loc[df["borough"] == borough]["borough"].count()
    sum_selss_by_borough = df.loc[df["borough"] == borough, "sale_price"].sum()
    # Hitung revenue per borough
    revenue_per_borough = sum_selss_by_borough / boroughes
    return boroughes, revenue_per_borough

# Hitung jumlah penjualan dan revenue per borough
borouth_satu = get_sale_price_by_borough(1)
borouth_dua = get_sale_price_by_borough(2)
borouth_tiga = get_sale_price_by_borough(3)
borouth_empat = get_sale_price_by_borough(4)
borouth_lima = get_sale_price_by_borough(5)

# Print hasil
print("JUMLAH SALE BOROUGH & JUMLAH REVENUE \n")
print("SEBELAH KIRI : borough, SEBELAH KANAN : revenue per borough")
print("")
print("borough 1")
print(borouth_satu)
print("")
print("borough 2")
print(borouth_dua)
print("")
print("borough 3")
print(borouth_tiga)
print("")
print("borough 4")
print(borouth_empat)
print("")
print("borough 5")
print(borouth_lima)
```

```python
# RATA RATA REVENUE TIAP PER QUARTAL
import pandas as pd

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/FIX NYC-PROPERTY-SALES.xlsx")

def average_revenue_by_quartal(start_date, end_date):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    data = df.query("sale_date.between(@start_date, @end_date)")
    return data["sale_price"].mean()

september = average_revenue_by_quartal("2016-09-01", "2016-09-30")
oktober = average_revenue_by_quartal("2016-10-01", "2016-10-31")
november = average_revenue_by_quartal("2016-11-01", "2016-11-30")
desember = average_revenue_by_quartal("2016-12-01", "2016-12-31")
januari = average_revenue_by_quartal("2017-01-01", "2017-01-31")
februari = average_revenue_by_quartal("2017-02-01", "2016-02-28")
maret = average_revenue_by_quartal("2017-03-01", "2017-03-31")
april = average_revenue_by_quartal("2017-04-01", "2017-04-30")
mei = average_revenue_by_quartal("2017-05-01", "2017-05-31")
juni = average_revenue_by_quartal("2017-06-01", "2017-06-30")
juli = average_revenue_by_quartal("2017-07-01", "2017-07-31")
agustus = average_revenue_by_quartal("2017-08-01", "2017-07-31")

# Cetak `sale_price_count`
print("RATA-RATA REVENUE TIAP BULAN\n")

print("\nSEPTEMBER")
print(september)
print("\nOktober")
print(oktober)
print("\nnovember")
print(november)
print("\ndesember")
print(desember)
print("\njanuari")
print(januari)
print("\nfebruari")
print(februari)
print("\nmaret")
print(maret)
print("\napril")
print(april)
print("\nmei")
print(mei)
print("\njuni")
print(juni)
print("\nJULI")
print(juli)
print('\nAGUSTUS')
print(agustus)
```
